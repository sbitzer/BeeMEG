## Correlation with momentary evidence (dot x-position)

```python
plot_absmax_channel('dot_x', 170)
```

Shown are the fitted beta-values of the momentary evidence regressor in a larger GLM which investigates effects of individual dots. A beta-value of 0.1 means that a standard unit change in dot x-position will result in a 0.1 standard unit change in the MEG signal of the selected channel while all other effects are held constant. The beta-value can, therefore, be roughly interpreted as the correlation of the MEG-signal with dot x-position.

Time 0 here represents the time after the onset of a dot. The effect is shown for the magnetometer channel which has the largest effect at 170 ms after dot onset (MEG2021). 

Each grey line is the result for a single subject. The black line shows the average effect over subjects. The 3 dotted black lines show inferred mean parameter values for when the trials are randomly permuted (3 different perumutations) so that the association between dot x-position and MEG signal is lost. The dashed black line shows, for comparison, the corresponding event-related field of that channel.

### Topographies at times of largest effect
While the figure above shows the size of the effect for a particular channel, the line below shows the maximum of the absolute mean (over subjects) beta value across channels, i.e., it shows the absolute size of the largest effect across channels at a particular time after dot onset. The figure below this line plot shows the topographies of magnetometers at the identified peaks of the line plot.

```python
show_topography_at_peaks('dot_x', order=3)
```

## Correlation with accumulated evidence (sum of previous dot x-positions)

```python
plot_absmax_channel('accev', 230)
```

Format of the figure is as above, but here the channel (MEG0731) has been selected based on the maximum effect of the accumulated evidence regressor for which results are shown. Note that positive accumulated evidence is evidence in favour of a decision for 'left' while positive dot x-position is evidence in favour of a decision for 'right'. Therefore, the negative effect here and the positive effect for dot x-position both signal a positive correlation of the MEG-signal with evidence in favour of a decision for 'right'.
