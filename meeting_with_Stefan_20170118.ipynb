# Basic regression of 'support correct' on MEG

Linear regression:
$$y = \beta \cdot x + \beta_0$$
with $x$ being support correct.

Using spatio-temporal cluster-level permutation analysis to compute multiple-comparison corrected p-values of $\beta$.

## p-values of largest cluster in data for all subjects

```python
subs = helpers.find_available_subjects(megdatadir='data/meg_final_data')
with np.load('data/inf_results/clusres_5thdot_meg_201701171125.npz') as res:
    max_cluster_p = pd.Series(res['max_cluster_p'], name='p of largest cluster',
                              index=pd.Index(subs, name='subject'))

sns.plt.figure()
sns.distplot(max_cluster_p, kde=False, rug=True);
```

# Regression of sign of support correct

```python
subs = helpers.find_available_subjects(megdatadir='data/meg_final_data')
with np.load('data/inf_results/clusres_5thdot_meg_201701171447.npz') as res:
    max_cluster_p_sign = pd.Series(res['max_cluster_p'], name='p of largest cluster',
                                   index=pd.Index(subs, name='subject'))

sns.plt.figure()
sns.distplot(max_cluster_p_sign, kde=False, rug=True);
```
